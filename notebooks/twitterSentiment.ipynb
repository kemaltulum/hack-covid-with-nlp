```python
sentiment_df = sentiment_df[sentiment_df['synonyms'].notna()]
```

```python
sentiment_df.shape
```

```python
senti_dict = {}

def get_features(row):
    d = {}
    
    d['label'] = row['Polarity Label']
    d['pos'] = row['pos value']
    d['neg'] = row['neg value']
    d['obj'] = row['obj value']

    return d

for ii in range(len(sentiment_df)):
    row = sentiment_df.iloc[ii]
    features = get_features(row)
    syns = str(row['synonyms'])
    if "," not in syns:
        senti_dict[syns.strip()] = features
    else:
        for syn in syns.split(','):
            senti_dict[syn.strip()] = features
```
